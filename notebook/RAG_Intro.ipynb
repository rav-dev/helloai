# RAG INTRO 

- RAG helps the LLM model to hallucinate less and solves the knowledge cut off problem of the LLMs
- It adds more knowledge and enhances the context of the LLM
- When model does not something we can provide additional documents to add context and send it to the model via our prompt 
- one disadvantage of this is that the model answer is limited to the limited amount of documents that we have provided


```python
import os
from dotenv import load_dotenv #to load the env vars securily
from openai import OpenAI #to create the openai client in order to communicate with the openai endpoints
```

```python
# Load variables from .env file
load_dotenv()
```

```python
# Access the keys
openai_key = os.getenv("OPENAI_API_KEY")
google_key = os.getenv("GOOGLE_API_KEY")

print("Keys loaded successfully")
```

- Now we will create an OPENAI client object that enbles us to connect to the openai endpoint

```python
#openai_client 
client = OpenAI()
```

- now we will create the user prompt 

```python
prompt = f"How many parameters LLaMA 4 Models have?"
```

```python
#now we will call the openai api end point using the client 
response = client.responses.create(
    model="gpt-5-chat-latest",
    temperature=0.0,
    input=prompt
)
```

- if you see the response of the gpt-5-chat-latest model it surely displays the knowledge cut around 2024. 
- LLaMA 4 has been released long time ago and surely our model isnt aware of it. 
- This time it has denied that it does not know  about LLaMA 4 but there might be instance where the model may hallucinate and produce result out of itself which is surely not correct

```python
#getting to know the response of the gpt-5-chat-latest model 
#hardcoded nicht so gut 
#the split() used at the end will remove additional whitespaces or special characters from both the start and end of the sequence.
print(response.output[0].content[0].text.strip())
```
